## 2. Load Data & Explore

```python
# Import packages
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load data from url
df = pd.read_csv('https://pasta.lternet.edu/package/data/eml/knb-lter-hfr/71/23/ab0fe2bf4f3ad850371ccb9c69d78469')
```

```python
df.info()
```

## 4. Data Wrangling

```python
# Subset to category 5 hurricanes
df_5 = df[df['ss'] == 5].copy()
```

```python
# Change startdate column to datetime
df['start.date'] = pd.to_datetime(df['start.date'])
```

```python
fig, ax = plt.subplots(figsize = (12, 5))

plt.scatter(df['start.date'], df['ss'], 
            c = '#3492eb',
            s = df['ss'] * 10)
plt.xlabel("Start Date")
plt.ylabel("Saffir-Simpson Hurricane Categories")
plt.yticks([1, 2, 3, 4, 5])
plt.title("Yucatan Peninsula: Hurricanes over Time")

# Display the graph
plt.show()
```
